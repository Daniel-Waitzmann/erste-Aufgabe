Hier werden Bibliotheken importiert.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Hier wird die Print-Funktion getested.

```python
print('Moin!')
```

Hier wird ein array erzeugt mit der arange-Funktion

```python
x = np.arange(-10,11)
```

Hier wird ein Plot erzeugt mit Titel und Achsenbeschriftung.

```python
plt.figure(figsize = (14,13))
plt.plot(x,x**2, color = 'r',linewidth = 5, linestyle = '-.')
plt.title('$Parable$', color = 'b', fontsize = 40, pad = 30)
plt.xlabel('x-axes', fontsize = 20, color = 'g')
plt.ylabel('y-axes', fontsize = 20, color = 'g');
```
